```python
def convert_timestamp_to_numeric_aqd(timestamp_string):
    new_string = timestamp_string[9:11]  + timestamp_string[12:14]+ timestamp_string[15:17]
    new_string = int(new_string)
    return new_string
```

```python
air_quality_data['Timestamp_numeric'] = air_quality_data['Timestamp'].apply(convert_timestamp_to_numeric_aqd)
```
