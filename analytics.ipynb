```python
def plot_df_merge_3(x1, y2, y3, y4, y5):
    plt.figure(figsize=(60,8))
    
    fig, ax1 = plt.subplots()
    
    
    ax1.set_xlabel('Date', fontsize = 15)
    ax1.set_ylabel('Temperature', color='tab:blue', fontsize = 15)
    ax1.plot(x1, y2, color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')
    
#     ax1.set_xticks(x1[::30])
    ax1.set_xticklabels(x1[::1], rotation=90, fontsize = 15)
    
    ax1.figure.set_figwidth(20)
    ax1.figure.set_figheight(5)
    
    ax2 = ax1.twinx()
    
    ax2.set_ylabel('CO2 Level', color='tab:green', fontsize = 15)  
    ax2.plot(x1, y3, color='tab:green')
    ax2.tick_params(axis='y', labelcolor='tab:green')
    
    ax2.figure.set_figwidth(20)
    ax2.figure.set_figheight(5)
    
    
    ax3 = ax1.twinx()
    
    ax3.spines.right.set_position(("axes", 1.05))
    
#     new_fixed_axis = ax3.get_grid_helper().new_fixed_axis
#     ax3.axis["right"] = new_fixed_axis(loc="right",
#                                     axes=ax3,
#                                     offset=(60, 0))
    
    ax3.set_ylabel('Humidity', color='tab:red', fontsize = 15)  
    ax3.plot(x1, y4, color='tab:red')
    ax3.tick_params(axis='y', labelcolor='tab:red')
    
    ax3.figure.set_figwidth(20)
    ax3.figure.set_figheight(5)

    
    ax4 = ax1.twinx()
    
    ax4.set_ylabel('Dust', color='tab:brown', fontsize = 15)  
    ax4.plot(x1, y5, color='tab:brown')
    ax4.tick_params(axis='y', labelcolor='tab:brown')
    
    ax4.figure.set_figwidth(10)
    ax4.figure.set_figheight(5)
#     plt.plot(x1, y3, color='tab:green')
#     plt.plot(x1, y4, color='tab:red')

#     plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
#     plt.figure().set_figwidth(210)
#     plt.figure().set_figheight(1)
#     plt.figure(figsize=(20,5))

    spine_distances = [0.0, 1.1, 1.2, 1.3]
    ax2.spines.right.set_position(('axes', spine_distances[1]))
    ax3.spines.right.set_position(('axes', spine_distances[2]))
    ax4.spines.right.set_position(('axes', spine_distances[3]))
    plt.plot()
```

## Reading and preprocessing occupant feedback data data 

```python
occupant_feedback_data = pd.read_csv('Occupant feedback thermal IAQ.csv', parse_dates=True)
```

```python
occupant_feedback_data.head()
```

```python
numeric_cols = ['Temperature [Overall]', 'Temperature [Head]', 'Temperature [Body]','Temperature [Hand]', 'Temperature [Leg]', 'Air Movement [Overall]', 'Air Movement [Head]', 'Air Movement [Body]', 'Air Movement [Hand]','How satisfied are you with the following: [Air quality]', 'Physical effects [Headaches]', 'Physical effects [Difficulty in concentration]', 'Physical effects [Dry skin]', 'Physical effects [Sleepiness]', 'Physical effects [Itching eyelids]', 'Physical effects [Dry eyes]', 'Physical effects [Throat irritation]', 'Physical effects [Runny nose]', 'Physical effects [Sneezing/coughing]']
```

```python
min_values = occupant_feedback_data[numeric_cols].min()
max_values = occupant_feedback_data[numeric_cols].max()
```

```python
normalized_data = ((occupant_feedback_data[numeric_cols]) / (max_values))*5
```

```python
normalized_data = pd.concat([normalized_data, occupant_feedback_data['Timestamp']], axis=1)
```

```python
normalized_data.head(282)
```

```python
# normalized_data['Timestamp'] = normalized_data['Timestamp'].astype(str)
```

```python
from datetime import datetime

def round_timestamp(time_str):
    time_str = str(time_str)
    time_str = time_str.strip()
    time_obj = datetime.strptime(time_str, '%H:%M')
    rounded_minutes = (time_obj.minute // 15) * 15
    rounded_time_obj = time_obj.replace(minute=rounded_minutes)
    rounded_time_str = rounded_time_obj.strftime('%H:%M')
    return rounded_time_str

```

```python
normalized_data = normalized_data.iloc[:282]
```

```python
normalized_data.head(300)
```

```python
normalized_data['timestamp_round'] = normalized_data['Timestamp'].apply(round_timestamp)
```

```python
def plot_2(x1, x2, y1, y2,y_label,x_label):
    plt.figure(figsize=(20,8))
    
    fig, ax1 = plt.subplots()
    
    ax1.set_ylabel(y_label, color='tab:green', fontsize = 15)  
    ax1.scatter(x2, y2, color='tab:green', alpha=0.05)
#     ax2.tick_params(axis='y', labelcolor='tab:green')
    
    ax1.figure.set_figwidth(20)
    ax1.figure.set_figheight(5)
    
#     ax1.set_xlim([min(min(x1), min(x2)), max(max(x1), max(x2))])
#     ax2.set_xlim([min(min(x1), min(x2)), max(max(x1), max(x2))])
    ax1.spines.right.set_position(('axes', -0.1))
    
    ax2 = ax1.twinx()
    
    
    ax2.set_xlabel(x_label, fontsize = 15)
    ax2.set_ylabel('Humidity', color='tab:blue', fontsize = 15)
    ax2.plot(x1, y1, color='tab:blue')
    ax2.tick_params(axis='y', labelcolor='tab:blue')
    
    ax2.set_xticklabels(x1[::1], rotation=90, fontsize = 15)
    
    ax2.figure.set_figwidth(20)
    ax2.figure.set_figheight(5)
    
    
    
    
    plt.plot()
```

```python
plot_2(final_dataset.index, normalized_data.index, final_dataset['Humidity'], normalized_data['Temperature [Overall]'],'Temperature Overall', 'Humidity')
```
